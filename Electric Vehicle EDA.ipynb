###### Model

```python
df['Model'].isna().sum()
```

```python
# Filling the missing values of model using groupby on other columns 
df['Model'] = df.groupby(['Make', 'Model Year', 'Electric Vehicle Type'])['Model'].transform(lambda x:x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))
```

```python
ev_count_by_pincode = df.groupby(['Postal Code', 'Model Year', 'State']).size().reset_index(name='Number_of_EV_Vehicles')
```

```python
state_data = ev_count_by_pincode[ev_count_by_pincode['State'] == 'WA']
```

```python
import plotly.express as px
```

```python
fig = px.choropleth_mapbox(state_data,
                           geojson="https://raw.githubusercontent.com/OpenDataDE/State-zip-code-GeoJSON/master/wa_washington_zip_codes_geo.min.json",
                           locations='Postal Code',
                           color='Number_of_EV_Vehicles',
                           featureidkey="properties.ZCTA5CE10",  
                           mapbox_style="carto-positron",
                           zoom=5,
                           center={"lat": 47.7511, "lon": -120.7401}, 
                           title="Number of EV vehicles based on location Washington Over Time",
                           animation_frame="Model Year",  
                           color_continuous_scale="Cividis",
                           hover_data=['Number_of_EV_Vehicles']  
                          )

# layoyt settings
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
df1 = df.groupby(['Make', 'Model Year']).size().reset_index(name='Number_of_Vehicles')
```

```python
import plotly.express as px
fig = px.bar(df1,
             y='Make',
             x='Number_of_Vehicles',
             animation_frame='Model Year',
             orientation='h',
             title='EV Makes and their Count Over the Years',
             labels={'Number_of_Vehicles': 'Number of EV Vehicles'},
             range_x=[0, 3000],
             color='Make', 
             color_discrete_map={
                 'Tesla': 'grey',
                 'Toyota': 'red',
                 'Ford': 'black',
             }
             )

top_year = df1['Model Year'].max()
if (df1['Model Year'] == top_year).any():
    make_2023 = df.loc[df['Model Year'] == top_year, 'Make'].iloc[0]
    fig.add_annotation(x=2500, y=make_2023,
                       text=f"Most EVs: {top_year}",
                       showarrow=False,
                       font_size=18)
else:
    print("Year 2023 isn't in the data")

# Customize the layout
fig.update_layout(
    xaxis=dict(showgrid=True, gridcolor='LightGray'),
    yaxis_title='EV Makes',
    xaxis_title='Number of EV Vehicles',
    showlegend=False,
    title_x=0.5,
    title_font=dict(size=20),
    margin=dict(l=50, r=50, t=50, b=50),
    width=800,
    height=600
)

#trace appearance
fig.update_traces(texttemplate='%{x}',
                  textposition='outside',
                  textfont_size=16)
fig.show()
```
